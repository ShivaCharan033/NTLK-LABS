```python
import os 
os.getcwd()
path = '/Users/charan/Desktop/Lab 1'
#1.1 Extracting Data
cleaned_data = []
with open("/Users/charan/Desktop/Lab 1/American-English.tsv", "r", encoding="utf-8") as dataset:
    for row in dataset:
        cleaned_row = row.strip()  
        print(cleaned_row)  
        cleaned_data.append(cleaned_row)  
print("Total rows extracted:", len(cleaned_data))
```

```python
#1.2 Remove Punctuation
import string
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
cleaned_data_no_punct = [remove_punctuation(row) for row in cleaned_data]
for row in cleaned_data_no_punct[:1069]:
    print(row)
```

```python
#1.3 Tokenization
from nltk.tokenize import word_tokenize
tokenized_data = [word_tokenize(row) for row in cleaned_data_no_punct]
for row in tokenized_data[:5]:
    print(row)
```

```python
#1.4 LowerCase
lowercase_data = [row.lower() for row in cleaned_data_no_punct]
for row in lowercase_data[:1069]:
    print(row)
```

```python
#1.5 Remove StopWards
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
data_no_stopwords = [[word for word in row if word.lower() not in stop_words] for row in tokenized_data]
for row in data_no_stopwords[:1069]:
    print(row)
```

```python
#1.6 Remove Numbers
data_no_numbers = [row[1:] if row[0].isdigit() else row for row in data_no_stopwords]

# Print the first five rows to check the result
for row in data_no_numbers[:1069]:
    print(row)
```

```python
#1.7 Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_data = [[lemmatizer.lemmatize(word) for word in row] for row in data_no_numbers]
for row in lemmatized_data[:1069]:
    print(row)
```

```python
#1.8 Word Frequency
from nltk import FreqDist
all_words = [word for row in lemmatized_data for word in row]
word_frequencies = FreqDist(all_words)
top_20_common_words = word_frequencies.most_common(20)
print(top_20_common_words)
```

# Second Exercise - Corpus Analysis and Data Visualization 

```python

#2.1 Matplot
import matplotlib.pyplot as plt
words, frequencies = zip(*top_20_common_words)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Frequencies')
plt.title('Top 20 Most Frequent Words')
plt.xticks(rotation=45)
plt.show()
```
